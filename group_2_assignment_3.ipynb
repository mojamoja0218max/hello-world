```
import urllib.request
from lxml import objectify, etree
```

```
def getYR(year):
  xml = urllib.request.urlopen('http://data.treasury.gov/feed.svc/DailyTreasuryYieldCurveRateData?$filter=year(NEW_DATE)%20eq%20'+year).read()
  tree = etree.fromstring(xml)
  tree.xpath("//*")
  
  def storeL(types):
    tb = tree.findall(".//{http://schemas.microsoft.com/ado/2007/08/dataservices}"+types)
    return tb

  def toText(mt):
    tb = [None] * len(mt)
    for i in range(len(mt)):
      tb[i]= mt[i].text
    return tb
    
  import numpy
  types = ['BC_1MONTH','BC_3MONTH','BC_6MONTH','BC_1YEAR','BC_2YEAR','BC_3YEAR','BC_5YEAR','BC_7YEAR','BC_10YEAR','BC_20YEAR','BC_30YEAR']
  grid = [None] * len(types)
  for i in range(len(types)):
    grid[i] = toText(storeL(types[i]))
  return(numpy.transpose(grid))
```

> The above code is a user-defined function, with string input (certain year) and output its 1 month, 6 month, etc. up to 30 year yield rates. It first sends requests to US Treasury website to retrieves the response.


> The 'storeL' is a user-defined function, which extracts and stores all the elements of certain type, and then returns the list.


> The 'toText' is a user-defined function, which convert each element into text form, and then returns the list.

> Lastly, it transposes the nested list so that each column is a yield rate type and each row is a date, and then it return the result.

The following code cell provides an example using this function, which also answers question 1.
